## Viewing Data Notebook

This notebook is to let us quickly look thru the trust in government data to see if there are any interesting trends for us to focus on for the task. Apologies for the scuffed-ness of some of it lol

-----------------------------------------------------------------

There are **two** viewing functions you can use here:
- plot_responses_collapsible lets you view one sampling of the data 
- compare_groups_collapsible lets you compare samplings of the data along one axis

In both cases, you have the option to view normalised data (to account for the different numbers of responses between demographic groups and between years), and collapsed data (either 'conviction' or 'sentiment'), which combines responses along agreement strength (strongly agree/disagree, agree/disagree, neutral) and agreement level (positive, negative, neutral). 

I've provided some examples of using each function to show you what I mean. E.g.:

    plot_responses_collapsible(df_test, filters={"Gender": "Male", "Region": "Regional", "Age": "18-34"}, question=5, normalise=True)

Will plot all the responses over time given by regional 18-34 years-old men, to question 5, with the responses normalised (to account for different amounts of responses in different years)

-----------------------------------------------------------------

**Possible values for Age:**
- 18-34
- 35-55
- 55+

**Possible values for Gender:**
- Female
- Male

**Possible values for Region:**
- Metro
- Regional

**Possible values for normalise:**
- True
- False (default if you don't write it in)

**Possible values for collapse:**
- None (default if you don't write it in)
- 'conviction' (strong, medium, neutral)
- 'sentiment' (positive, negative, neutral)

**Possible values for question:**
- 1 -> 15
